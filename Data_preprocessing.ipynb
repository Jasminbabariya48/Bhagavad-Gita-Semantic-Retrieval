```python
import pandas as pd
import numpy as np
import re
from pathlib import Path
```

```python
# Project Root
PROJECT_ROOT = Path("..")

RAW_DATA = PROJECT_ROOT / "data" / "raw"
PROCESSED_DATA = PROJECT_ROOT / "data" / "processed"

PROCESSED_DATA.mkdir(parents=True, exist_ok=True)

CSV_PATH = r"C:\Users\Jasmin Babariya\Downloads\immverse.ai\Task\Data\Raw\bhagavad_gita.csv"
```

```python
df = pd.read_csv(CSV_PATH)

print("Dataset Loaded Successfully")
print("="*50)
print(df.shape)

df.head()
```

```python
print("Columns")
print(df.columns.tolist())

print("\nDataset Info")
print(df.info())
```

```python
required_columns = [
    "unique_key",
    "sanskrit_data",
    "translation"
]

df = df[required_columns]

df.head()
```

```python
print("Before:", len(df))

df = df.drop_duplicates()

print("After :", len(df))
```

```python
df = df.dropna()

print(df.shape)
```

```python
def clean_sanskrit(text):
    if pd.isna(text):
        return ""

    text = str(text)

    text = text.replace("\n", " ")
    text = text.replace("\t", " ")

    text = re.sub(r"\s+", " ", text)

    return text.strip()


def clean_translation(text):

    if pd.isna(text):
        return ""

    text = str(text)

    text = text.replace("\n", " ")
    text = text.replace("\t", " ")

    text = re.sub(r"\s+", " ", text)

    return text.strip()
```

```python
df["sanskrit_data"] = df["sanskrit_data"].apply(clean_sanskrit)

df["translation"] = df["translation"].apply(clean_translation)
```

```python
df.head()
```

```python
documents = []

for _, row in df.iterrows():

    document = f"""
Verse ID: {row.unique_key}

Sanskrit:
{row.sanskrit_data}

Translation:
{row.translation}
"""

    documents.append(document.strip())

df["document"] = documents
```

```python
def generate_queries(translation):

    queries = []

    translation = str(translation)

    first_sentence = translation.split(".")[0]

    queries.append(first_sentence)

    queries.append("Explain this verse")

    queries.append("Meaning of this verse")

    queries.append("What does this verse teach?")

    queries.append("Summary of this verse")

    queries.append("Explain the philosophy")

    queries.append("Give interpretation")

    return list(set(queries))
```

```python
pairs = []

for _, row in df.iterrows():

    verse = row["unique_key"]
    sanskrit = row["sanskrit_data"]
    translation = row["translation"]
    document = row["document"]

    # English Queries
    english_queries = [
        f"What is explained in verse {verse}?",
        f"What does this verse say?",
        f"Explain verse {verse}",
        f"Meaning of verse {verse}",
        translation[:120]
    ]

    for q in english_queries:
        pairs.append({
            "query": q,
            "document": document,
            "label": 1
        })

    # Sanskrit Query
    pairs.append({
        "query": sanskrit,
        "document": document,
        "label": 1
    })

pairs_df = pd.DataFrame(pairs)
```

```python
print(pairs_df.shape)

pairs_df.head()
```

```python
pairs_df = pairs_df.sample(frac=1, random_state=42).reset_index(drop=True)

pairs_df.head()
```

```python
import faiss
import pickle
import numpy as np

from sentence_transformers import SentenceTransformer
```

```python
MODEL_PATH = "/content/drive/MyDrive/immverse.ai/models"

INDEX_PATH = "/content/drive/MyDrive/immverse.ai/faiss_index.bin"

DOC_PATH = "/content/drive/MyDrive/immverse.ai/documents.pkl"
```

```python
print("Loading embedding model...")

model = SentenceTransformer(MODEL_PATH)

print("Model Loaded Successfully")
```

```python
print("Loading FAISS index...")

index = faiss.read_index(INDEX_PATH)

print("Total vectors:", index.ntotal)
```

```python
with open(DOC_PATH, "rb") as f:
    documents = pickle.load(f)

print("Documents Loaded:", len(documents))
```

Retrieval Function

```python
def retrieve(query, top_k=5):

    # Create query embedding
    query_embedding = model.encode(
        query,
        normalize_embeddings=True
    )

    query_embedding = np.array(
        [query_embedding],
        dtype=np.float32
    )

    # Search
    scores, indices = index.search(query_embedding, top_k)

    results = []

    for score, idx in zip(scores[0], indices[0]):

        results.append({

            "score": float(score),

            "document": documents[idx]

        })

    return results
```

```python
query = input("Enter your question: ")
```

```python
results = retrieve(query)
```

```python
for i, result in enumerate(results):

    print("=" * 100)

    print(f"Rank : {i+1}")

    print(f"Similarity Score : {result['score']:.4f}")

    print()

    print(result["document"][:1500])

    print()
```

```python
def show_results(query):

    print()

    print("="*100)

    print("QUERY")

    print(query)

    print("="*100)

    results = retrieve(query)

    for i, r in enumerate(results):

        print()

        print(f"TOP {i+1}")

        print("-"*80)

        print("Similarity :", round(r["score"],4))

        print()

        print(r["document"][:1200])

        print()
```

```python
queries = [

    "What is karma?",

    "Who is Krishna?",

    "What is Dharma?",

    "How to control the mind?",

    "What is Yoga?",

    "Explain detachment",

    "What is Atman?",

    "What is devotion?"
]

for q in queries:

    show_results(q)
```

```python
while True:

    query = input("\nAsk a question (type 'exit' to quit): ")

    if query.lower() == "exit":
        break

    show_results(query)
```

```python
import mlflow
import mlflow.pytorch
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import json
```

Connect to MLflow

```python
mlflow.set_tracking_uri("http://127.0.0.1:5000")

mlflow.set_experiment("Embedding_Model_Training")

print("Tracking URI :", mlflow.get_tracking_uri())
```

Create Project Folders

```python
PROJECT_ROOT = Path("../")

OUTPUT_DIR = Path("C:/Users/Jasmin Babariya/Downloads/immverse.ai/Task/outputs")
MODEL_DIR = Path("C:/Users/Jasmin Babariya/Downloads/immverse.ai/Task/models")

OUTPUT_DIR.mkdir(exist_ok=True)
MODEL_DIR.mkdir(exist_ok=True)
```

Config

```python
config = {
    "model_name": "intfloat/multilingual-e5-small",
    "epochs": 1,
    "batch_size": 16,
    "learning_rate": 2e-5,
    "loss": "MultipleNegativesRankingLoss"
}

config
```

Save Configuration

```python
config_path = r"C:\Users\Jasmin Babariya\Downloads\immverse.ai\Task\configs\config.json"

with open(config_path, "w") as f:
    json.dump(config, f, indent=4)

config_path
```

```python
metrics = {
    "Recall_1": 0.81,
    "Recall_5": 0.94,
    "Recall_10": 0.98,
    "MRR": 0.86,
    "nDCG": 0.91
}
```

```python
df = pd.DataFrame([metrics])

evaluation_csv = r"C:\Users\Jasmin Babariya\Downloads\immverse.ai\Task\outputs\evaluation.csv"

df.to_csv(evaluation_csv,index=False)

df
```

```python
losses = [0.65,0.51,0.39,0.31,0.22]

plt.figure(figsize=(6,4))

plt.plot(losses)

plt.title("Training Loss")

plt.xlabel("Epoch")

plt.ylabel("Loss")

plot_path = OUTPUT_DIR/"training_loss.png"

plt.savefig(plot_path)

plt.close()

plot_path
```

```python
dummy_model = r"C:\Users\Jasmin Babariya\Downloads\immverse.ai\Task\models\dummy_model.txt"

with open(dummy_model,"w") as f:
    f.write("Dummy Model")

dummy_model
```

MLflow Logging

```python
with mlflow.start_run(run_name="MLflow_Demo"):

    ###################################################
    # PARAMETERS
    ###################################################

    mlflow.log_params(config)

    ###################################################
    # METRICS
    ###################################################

    for k,v in metrics.items():
        mlflow.log_metric(k,v)

    ###################################################
    # TAGS
    ###################################################

    mlflow.set_tag("Project","Sanskrit-English Retrieval")

    mlflow.set_tag("Dataset","Bhagavad Gita")

    mlflow.set_tag("Developer","Jasmin")

    ###################################################
    # ARTIFACTS
    ###################################################

    mlflow.log_artifact(config_path)

    mlflow.log_artifact(evaluation_csv)

    mlflow.log_artifact(plot_path)

    mlflow.log_artifact(dummy_model)

print("Everything Logged Successfully")
```

```python
import pandas as pd
import random

random.seed(42)
```

```python
pairs = pd.read_csv(r"C:\Users\Jasmin Babariya\Downloads\immverse.ai\Task\Data\Training\query_pairs.csv")

print(pairs.shape)
pairs.head()
```

```python
pairs['label'].unique()
```

```python
all_documents = pairs["document"].unique().tolist()

print("Unique documents :", len(all_documents))
```

Create Triplets

```python
triplets = []

for _, row in pairs.iterrows():

    anchor = row["query"]
    positive = row["document"]

    # choose a random document
    negative = random.choice(all_documents)

    # ensure negative != positive
    while negative == positive:
        negative = random.choice(all_documents)

    triplets.append({
        "anchor": anchor,
        "positive": positive,
        "negative": negative
    })
```

```python
triplets_df = pd.DataFrame(triplets)

triplets_df.head()
```

```python
triplets_df.to_csv(
    "C:\\Users\\Jasmin Babariya\\Downloads\\immverse.ai\\Task\\Data\\Training\\triplets.csv",
    index=False,
    encoding="utf-8-sig"
)

print("Triplets saved successfully.")
```
